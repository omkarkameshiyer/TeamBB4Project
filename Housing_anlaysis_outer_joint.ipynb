```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.ioff()  
```

# Cannabis data

```python
cannabis_df = pd.read_csv("TeamBB4Project/CannabisLicData_cleaned.csv")
cannabis_df
```

```python
def lic_type_adult(x):
    if x == "BOTH" or x == "Adult-Use":
        return 1
    else:
        return 0

def lic_type_Medicinal(x):
    if x == "BOTH" or x == "Medicinal":
        return 1
    else:
        return 0
    
cannabis_df["Lic Type: Adult-Use"] = cannabis_df["Adult-Use/Medicinal"].apply(lic_type_adult)
cannabis_df["Lic Type: Medicinal"] = cannabis_df["Adult-Use/Medicinal"].apply(lic_type_Medicinal)
cannabis_df
```

```python
cannabis_df.count()
```

```python
cannabis_df = cannabis_df.dropna(how='any')
cannabis_df.count()
```

```python
cannabis = cannabis_df.loc[:, ["County", "Zip Code", "Adult-Use/Medicinal"]]
cannabis = cannabis_df.rename(columns={"Adult-Use/Medicinal":"License Type"})
cannabis.head()
```

```python
la_cannabis = cannabis.loc[cannabis["County"] == "LOS ANGELES"][["County", "Zip Code"]]
la_cannabis_by_zip = la_cannabis.groupby(la_cannabis["Zip Code"]).count()
la_cannabis_by_zip = la_cannabis_by_zip.rename(columns={"County":"License Count"})
la_cannabis_by_zip.sort_values(["License Count"], ascending=False)
```

```python
la_cannabis_by_zip.sort_values(["License Count"], ascending=False).plot(kind="bar", color = "skyblue", figsize=(10,3))
plt.tight_layout()
plt.show()
```

# Housing data

```python
housing_df = pd.read_csv("TeamBB4Project/LA_RealEstateMarketOverview_1yr_cleaned.csv")
housing_df = housing_df.rename(columns={"Region Name":"Zip Code"})
housing_df.head()
```

# Merged data

```python
merged_df = pd.merge(la_cannabis_by_zip, housing_df, on="Zip Code", how="outer")
merged_df = merged_df.fillna(0)
merged_df = merged_df.set_index(["Zip Code"])
merged_df.sort_values(["License Count"], ascending=False)
```

```python
merged_df["Zillow Home Value Index_Current"].replace('---', np.nan, inplace=True)
merged_df = merged_df.dropna(how='any')
merged_df.count()
```

```python
merged_df["Zillow Home Value Index_Current"] = merged_df["Zillow Home Value Index_Current"].replace('[\$,]', '', regex=True).astype(int)
merged_df["Zillow Rent Index_Current"] = merged_df["Zillow Rent Index_Current"].replace('[\$,]', '', regex=True).astype(int)
merged_df["Zillow Home Value Index_Year Over Year"] = merged_df["Zillow Home Value Index_Year Over Year"].replace('[\%,]', '', regex=True).astype(float)
merged_df["Zillow Rent Index_Year Over Year"] = merged_df["Zillow Rent Index_Year Over Year"].replace('[\%,]', '', regex=True).astype(float)
```

```python
max_home_value = merged_df["Zillow Home Value Index_Current"].max()
max_rent_price = merged_df["Zillow Rent Index_Current"].max()
max_home_value, max_rent_price
```

```python
with_dispensary = merged_df.loc[merged_df["License Count"] == 0.0]
median_yoy_home_with_dispensary = with_dispensary["Zillow Home Value Index_Year Over Year"].median()
median_yoy_rent_with_dispensary = with_dispensary["Zillow Rent Index_Year Over Year"].median()
avg_yoy_home_with_dispensary = with_dispensary["Zillow Home Value Index_Year Over Year"].mean()
avg_yoy_rent_with_dispensary = with_dispensary["Zillow Rent Index_Year Over Year"].mean()

without_dispensary = merged_df.loc[merged_df["License Count"] > 0.0]
median_yoy_home_without_dispensary = without_dispensary["Zillow Home Value Index_Year Over Year"].median()
median_yoy_rent_without_dispensary = without_dispensary["Zillow Rent Index_Year Over Year"].median()
avg_yoy_home_without_dispensary = without_dispensary["Zillow Home Value Index_Year Over Year"].mean()
avg_yoy_rent_without_dispensary = without_dispensary["Zillow Rent Index_Year Over Year"].mean()

yoy_summary_home = pd.DataFrame({"with dispensary": [median_yoy_home_with_dispensary],
                            "without dispensary": [median_yoy_home_without_dispensary],
#                             "Avg Home Value Change YoY% (with dispensary)": [avg_yoy_home_with_dispensary],
#                             "Avg Home Value Change YoY% (without dispensary)": [avg_yoy_home_without_dispensary]
                           })
yoy_summary_home 
```

```python
# yoy_summary_home.plot(kind="bar", facecolor="skyblue")
yoy_summary_home.transpose().plot(kind="bar", facecolor="skyblue", rot=0)
plt.title("Median Home Value Change% YoY Comparison")
plt.tight_layout()
plt.show()
```

```python
yoy_summary_rent = pd.DataFrame({"with dispensary": [median_yoy_rent_with_dispensary],
                                 "without dispensary": [median_yoy_rent_without_dispensary]
#                                 "Avg Rent Price Change YoY (with dispensary)": [avg_yoy_rent_with_dispensary],
#                                "Avg Rent Price Change YoY (without dispensary)": [avg_yoy_rent_without_dispensary]
                           })
yoy_summary_rent 
```

```python
yoy_summary_rent.transpose().plot(kind="bar", facecolor="skyblue", rot=0)
plt.title("Median Rent Price Change% YoY Comparison")
plt.tight_layout()
plt.show()
```

```python
# Build the scatter plots for each zip code
area_scale = merged_df["License Count"]*80
home_price = plt.scatter(merged_df["Zillow Home Value Index_Current"], merged_df["Zillow Rent Index_Current"], marker="o", facecolors="skyblue", edgecolors="grey", s=area_scale, alpha=0.75)

# Incorporate the other graph properties
plt.title("Home Value vs. Rent Price")
plt.xlabel("Home Value$")
plt.ylabel("Rent Price$")
plt.tight_layout()

# Save Figure
plt.savefig("ScatterPlot1.png")

# Show plot
plt.show()
```

```python
# Build the scatter plots for each zip code
area_scale = merged_df["License Count"]*80
home_price = plt.scatter(merged_df["Zillow Home Value Index_Current"], merged_df["Zillow Home Value Index_Year Over Year"], marker="o", facecolors="skyblue", edgecolors="grey", s=area_scale, alpha=0.75)

# Incorporate the other graph properties
plt.title("Home Value vs. YoY Change")
plt.xlabel("Home Value$")
plt.ylabel("YoY Change%")
plt.tight_layout()

# Save Figure
plt.savefig("ScatterPlot2.png")

# Show plot
plt.show()
```
